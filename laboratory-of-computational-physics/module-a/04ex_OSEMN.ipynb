1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
import numpy.random as npr
npr.seed(123)

size = 10
data = npr.rand(size)
print("List of " + str(size) + " random numbers:\n", data)

file_name = "simple_data.txt"
with open(file_name, mode='w') as f:
    for i in range(len(data)):
        f.write(str(data[i]))
        if i < len(data)-1:   # to avoid writing "\n" after writing the last number of the list
            f.write("\n")

with open(file_name, mode='r') as f:
    print("\nContent of the file '%s':" % file_name)
    print(f.read())   # printing the whole file
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
N = 5
matrix = npr.rand(N,N)
print("%dx%d matrix of random numbers:\n" % (N,N), matrix)

file_name = "data.txt"
with open(file_name, mode='w') as f:
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            f.write(str(matrix[i][j]))
            if j < matrix.shape[1]-1:   # to avoid writing " " after writing the last element of the i-th row
                f.write(" ")
        if i < matrix.shape[0]-1:   # to avoid writing "\n" after writing the last element of the last row
            f.write("\n")

with open(file_name, mode='r') as f:
    print("\nContent of the file '%s':" % file_name)
    print(f.read())   # printing the whole file
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
import csv

txt_name = "data.txt"
data = np.loadtxt(txt_name)

csv_name = "data.csv"
with open(csv_name, 'w') as f:
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            f.write(str(data[i][j]))
            if j < data.shape[1]-1:   # to avoid writing "," after writing the last element of the i-th row
                f.write(",")
        if i < data.shape[0]-1:   # to avoid writing "\n" after writing the last element of the last row
            f.write("\n")

with open(csv_name, 'r') as f:
    print("Content of the file '%s':" % csv_name)
    for row in csv.reader(f):
        print(row)
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
file_name = "credit_card.dat"
with open(file_name, 'rb') as f:
    file_content = f.readlines()
    char_size = 6   # size of each character in bits
    print("Content of the file '%s':" % file_name)
    for line in file_content:
        card = ''
        for i in range(0, len(line), char_size):
            card += chr(int(line[i:i+char_size], 2))
        if len(card) == 20:   # to avoid printing meaningless lines
            print(card, end='')
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json

json_name = 'user_data.json'
data = json.load(open(json_name))

csv_name = 'user_data.csv'
with open(csv_name, 'w') as f:
    writer = csv.DictWriter(f, fieldnames = data[0], lineterminator='\n')
    writer.writeheader()
    for i in data:
        if i['CreditCardType'] == 'American Express':
                writer.writerow(i)

with open(csv_name, 'r') as f:
    print("Content of the file '%s':" % csv_name)
    for line in csv.reader(f):
        print(line)
```
