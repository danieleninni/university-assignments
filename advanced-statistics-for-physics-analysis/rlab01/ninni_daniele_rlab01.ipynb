## **Exercise 1** - Vectors and Data Frames

The following table gives the volume, area, length and maximum and mean depths of some Scottish lakes [[1]](#ref1). Create vectors, holding the lake's name and all the parameters and build a data frame called $\texttt{scottish.lakes}$ from the vectors.

```r
scottish.lakes <- read.table('scotland_lochs.txt', header=T, sep=',')
scottish.lakes
```

<a id='sec1.1'></a>
### **1.1** Evaluate the highest and lowest volume and area lake

```r
# sort data frame 'scottish.lakes' by column 'Volume_km3' in increasing order
scottish.lakes <- scottish.lakes[order(scottish.lakes$Volume_km3),]
cat('Highest volume lake:', tail(scottish.lakes, n=1)$Loch)
cat('\nLowest volume lake:', head(scottish.lakes, n=1)$Loch)

# sort data frame 'scottish.lakes' by column 'Area_km2' in increasing order
scottish.lakes <- scottish.lakes[order(scottish.lakes$Area_km2),]
cat('\n\nHighest area lake:', tail(scottish.lakes, n=1)$Loch)
cat('\nLowest area lake:', head(scottish.lakes, n=1)$Loch)
```

<a id='sec1.2'></a>
### **1.2** Order the frame with respect to the area and determine the two largest area lakes

```r
# sort data frame 'scottish.lakes' by column 'Area_km2' in decreasing order
scottish.lakes <- scottish.lakes[order(scottish.lakes$Area_km2, decreasing=T),]
scottish.lakes

cat('Largest area lake:', scottish.lakes$Loch[1])
cat('\nSecond largest area lake:', scottish.lakes$Loch[2])
```

<a id='sec1.3'></a>
### **1.3** By summing up the areas occupied by the lakes, determine the area of Scotland covered by water

```r
cat('Area of Scotland covered by water:', sum(scottish.lakes$Area_km2), 'km^2')
```

## **Exercise 2** - Crude Oil Production

The following CSV file https://drive.google.com/file/d/13WkinYhJNSr_XCldb3wpZOI0fgjer7Iw/view contains data on crude oil prices from 1861 to 2020, measured in US dollars per barrel [[2]](#ref2).

<a id='sec2.1'></a>
### **2.1** Write R code that is able to read the file and import it in a data frame structure (Hint: before loading the file, open it with a text editor and check its structure)

```r
df_oil <- read.table('crude-oil-prices.csv', header=T, sep=',')
colnames(df_oil) <- c('Entity', 'Code', 'Year', 'Price')
df_oil
```

<a id='sec2.2'></a>
### **2.2** Produce a plot with the oil price as a function of the year

```r
ggplot(data=df_oil, aes(x=Year, y=Price)) +
    geom_line(colour='red', size=.5) +
    geom_point(alpha=.5, colour='black', size=.5) +
    scale_x_continuous(n.breaks=17) +
    scale_y_continuous(n.breaks=12) +
    ylab('Oil price ($)') +
    ggtitle('Oil price as a function of the year')
```

<a id='sec2.3'></a>
### **2.3** Which is the highest price in history? When did it occur?

```r
cat('Highest price in history:', max(df_oil$Price), '$')
cat('\nYear corresponding to the highest price in history:', df_oil[which.max(df_oil$Price),'Year'])
```

<a id='sec2.4'></a>
### **2.4** Plot the derivative of the curve, simply evaluated with the finite difference formula (forward derivative):

$$
\frac{\partial price}{\partial year} = price_{j+1} - price_j
$$

```r
dPrice <- df_oil$Price[-1] - df_oil$Price[-nrow(df_oil)]

ggplot() +
    geom_line(aes(x=df_oil$Year[-nrow(df_oil)], y=dPrice), colour='green', size=.5) +
    geom_point(aes(x=df_oil$Year[-nrow(df_oil)], y=dPrice), alpha=.5, colour='black', size=.5) +
    scale_x_continuous(n.breaks=17) +
    scale_y_continuous(n.breaks=9) +
    xlab('Year') +
    ylab('dPrice') +
    ggtitle('Derivative of oil price w.r.t. year')
```

## **Exercise 3** - World Coal Production

The following CSV file https://drive.google.com/file/d/1U1hK5o_d3vl8twwGaRJNcDJ79O9I5zJ1/view?usp=sharing contains data on the coal production (in TW $\cdot$ hour) for several countries in the world [[3]](#ref3).

<a id='sec3.1'></a>
### **3.1** Write R code that is able to read the file and import it in a tibble [[4]](#ref4) structure

```r
tbl_coal <- read_csv('coal-production-by-country.csv')
tbl_coal
```

<a id='sec3.2'></a>
### **3.2** Count the number of countries available in the file and produce a barplot with the number of entries for each country

```r
tbl_coal_by_country <- tbl_coal %>%
                        group_by(Entity) %>% # group tibble 'tbl_coal' by column 'Entity'
                        filter(!(str_starts(Code, 'OWID_'))) %>% # retain only rows corresponding to countries
                        tally(name='Frequency') # count number of rows in each group
tbl_coal_by_country

cat('Number of countries available in the file:', nrow(tbl_coal_by_country))
```

```r
# for better visualization: multiple barplots representing different subsets of tibble 'tbl_coal_by_country'
n_barplots <- 9 # number of barplots
subset_nrow <- nrow(tbl_coal_by_country)/n_barplots # number of rows in each subset

for (i in 1:n_barplots) {
    print(
        ggplot(data=tbl_coal_by_country %>% slice((((i-1)*subset_nrow)+1):(i*subset_nrow)), aes(x=Frequency, y=Entity)) +
            geom_col(width=.5, alpha=.5, colour='black', fill='orange', size=.3) +
            scale_x_continuous(n.breaks=13) +
            ylab('Country') +
            ggtitle(paste('Bar chart (tbl_coal_by_country, subset #', i, ')', sep=''))
    )
}
```

### For the following items select only the years $\geq$ 1970

```r
tbl_coal_1970 <- tbl_coal %>% filter(Year >= 1970)
tbl_coal_1970
```

<a id='sec3.3'></a>
### **3.3** Selecting only the years after 1970, determine the total integrated production for each country and print the top 5 countries with highest coal production

```r
tbl_coal_1970_by_country <- tbl_coal_1970 %>%
                                group_by(Entity) %>% # group tibble 'tbl_coal_1970' by column 'Entity'
                                filter(!(str_starts(Code, 'OWID_'))) %>% # retain only rows corresponding to countries
                                summarise('Total coal production (TWh)'=sum(`Coal production (TWh)`)) %>% # sum along column 'Coal production (TWh)'
                                arrange(desc(`Total coal production (TWh)`)) # sort by column 'Total coal production (TWh)' in decreasing order
tbl_coal_1970_by_country
```

```r
cat('Top 5 countries with highest coal production:')
head(tbl_coal_1970_by_country, n=5)
```

<a id='sec3.4'></a>
### **3.4** For the top 5 countries, create a plot of production as a function of time

```r
top_5_countries <- c(head(tbl_coal_1970_by_country, n=5)$Entity)

ggplot(data=filter(tbl_coal_1970, Entity %in% top_5_countries), aes(x=Year, y=`Coal production (TWh)`, group=Entity, colour=Entity)) +
    geom_line(size=.5) +
    geom_point(size=.8) +
    scale_x_continuous(n.breaks=11) +
    scale_y_continuous(n.breaks=12) +
    labs(colour='Country') +
    ggtitle('Coal production as a function of time')
```

<a id='sec3.5'></a>
### **3.5** Generate a plot with the cumulative sum of the world coal production over the years

```r
ggplot(data=filter(tbl_coal_1970, Entity == 'World'), aes(x=Year, y=cumsum(`Coal production (TWh)`))) +
    geom_line(colour='black', size=.5) +
    geom_point(colour='red', size=1) +
    scale_x_continuous(n.breaks=9) +
    scale_y_continuous(n.breaks=8) +
    ylab('Coal production (TWh)') +
    ggtitle('Cumulative sum of the world coal production over the years')
```

## **Exercise 4** - COVID-19 Vaccine Data

Data on COVID-19 vaccine can be found on [[5]](#ref5).

The file $\texttt{vaccinations-by-manufacturer.csv}$ (https://github.com/owid/covid-19-data/blob/master/public/data/vaccinations/vaccinations-by-manufacturer.csv) contains a summary on the vaccination data grouped by vaccine type and country.

```r
tbl_vax_mfr <- read_csv('vaccinations-by-manufacturer.csv')
tbl_vax_mfr
```

```r
# to avoid plotting problems due to either missing data or wrong data, let's fix the tibble 'tbl_vax_mfr' so that:
#   for each 'location':
#     for each 'vaccine' used in 'location':
#       'total_vaccinations' increases over 'date'
#       for each 'date':
#         there is one and only one 'total_vaccinations'
tbl_vax_mfr <- tbl_vax_mfr %>%
                            group_by(location, vaccine) %>%
                            filter(diff(c(1, total_vaccinations)) >= 0) %>%
                            pivot_wider(names_from=vaccine, values_from=total_vaccinations) %>%
                            fill(everything()) %>%
                            pivot_longer(!c('location', 'date'), names_to='vaccine', values_to='total_vaccinations', values_drop_na=T)
tbl_vax_mfr
```

<a id='sec4.1'></a>
### **4.1** $\texttt{filter()}$ the original tibble by selecting the following countries: Italy

```r
tbl_italy <- filter(tbl_vax_mfr, location == 'Italy')
tbl_italy
```

<a id='sec4.2'></a>
### **4.2** Plot the number of vaccines given as a function of time for the different vaccine manufacturer

```r
ggplot(data=tbl_italy, aes(x=date, y=total_vaccinations, group=vaccine, colour=vaccine)) +
    geom_line(size=.5) +
    scale_x_date(date_breaks='1 month', date_labels='%Y-%m') +
    scale_y_continuous(n.breaks=10) +
    xlab('Date') +
    ylab('Total vaccinations') +
    labs(colour='Vaccine') +
    theme(axis.text.x = element_text(hjust=1, vjust=1, angle=30)) +
    ggtitle('Italy - Number of vaccines given as a function of time')
```

<a id='sec4.3'></a>
### **4.3** From the same tibble plot the total number of vaccines shot per day in Italy

```r
ggplot(data=tbl_italy %>% group_by(date) %>% summarise('total_vaccinations'=sum(total_vaccinations)), aes(x=date, y=c(NA, diff(total_vaccinations)))) +
    geom_line(na.rm=T, colour='blue', size=.3) +
    scale_x_date(date_breaks='1 month', date_labels='%Y-%m') +
    scale_y_continuous(n.breaks=8) +
    xlab('Date') +
    ylab('Daily vaccinations') +
    theme(axis.text.x = element_text(hjust=1, vjust=1, angle=30)) +
    ggtitle('Italy - Total number of vaccines shot per day')
```
